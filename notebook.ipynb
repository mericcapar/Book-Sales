```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Books_Data_Clean.csv")
```

```python
df
```

```python
df.head()
```

```python
df.describe().T
```

```python
df = df[df["Publishing Year"] > 1900] # Publishing Year -560 olan data oldugu icin boyle bir yol tercih ediyoruz
```

```python
df.describe().T
```

```python
df.isna().sum() #Null degerleri buluyoruz. 
```

```python
df.dropna(subset = "Book Name" , inplace= True) #Cok satan bir kitabin language_code u belli olmayabilir fakat Book name i olmayan bir kitap onemsizdir.
```

```python
df.isna().sum() #Kontrol ediyoruz.
```

```python
df.duplicated().sum() #Tekrarlayan veri var mi diye kontrol ediyoruz. 
```

```python
df.nunique() #Unique degerlere bakiyoruz. 
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing Year")
plt.show()
#We can see that after 20's there is more books published. 
```

```python
df["genre"].value_counts().plot(kind = "bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of books in each genre")
plt.show()
#Genre fiction more prefered than other genres.
```

```python
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending= False) #Yazarlarin puanlamalarini buyukten kucuge siraladik
```

```python
sns.boxplot(x = "genre", y= "Book_ratings_count" , data =df)
plt.xlabel("Genre")
plt.ylabel("Book ratings count")
plt.title("Box Plot of Book Ratings Count for Each Genre")
plt.show()
```

```python
plt.scatter(df["sale price"] , df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Unit Sold")
plt.title("Scatter Plot of Sale Price vs Unit Sold")
plt.show()  
#We can see that if prices increase units sold decreases
```

```python
df["language_code"].value_counts().sort_values(ascending= False) #Most used languages for books
```

```python
language_counts = df["language_code"].value_counts()
```

```python
plt.pie(language_counts , labels = language_counts.index , startangle=90 , autopct= "%1.1f%%") #with this Autopct, we can see percentages
plt.title("Language Distribution of Books")
plt.show()
```

```python
total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum() # We can see that how many books authors sell
```

```python
total_gross_sales_by_author.sort_values(ascending= False).head(10) 
```

```python
total_gross_sales_by_author.sort_values(ascending= False).head(15).plot(kind = "bar")
plt.xlabel("Author") , plt.ylabel("Total Gross Sales") , plt.title("Total gross sales for each Author")
plt.show()
```

```python
sns.boxplot(x = "Author_Rating" , y = "units sold" , data = df)
plt.xlabel("Author Rating") , plt.ylabel("Units Sold") , plt.title("Box Plot of Units Sold for Each Author Rating")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind = "line" , marker = ".")
plt.xlabel("Publishing Year") , plt.ylabel("Total Units Sold") , plt.title("Total Units Sold Over the Year")
plt.show()
#Total books sold over the year
```
